```python
import sqlite3
import csv
```

```python
#In memory database
dbconn = sqlite3.connect(':memory:')

# set a variable up for my cursor so I can write SQL scripts
db = dbconn.cursor()

# create Employees table
db.execute("""CREATE TABLE Employees (
            employee_id TEXT,
            name TEXT,
            address TEXT,
            hired_date TEXT,
            salary REAL
            )""")

#commits your transactions into the DATABASE
dbconn.commit()
```

```python
#transfers data from csv file of initial ten employees to an SQL table


db = dbconn.cursor()

try :
    with open("employee_copy.csv") as data_file :
        csv_data = csv.reader(data_file, delimiter = ',')

        for row in csv_data:
            employee = row[0],row[1],row[2],row[3],row[4]
            db.execute('''INSERT INTO Employees(employee_id,name,address,hired_date,salary) VALUES (?,?,?,?,?)''', (employee))

        
    
except :
    print("ERROR:employee.csv doesn't exist!")
    

        

dbconn.commit()
```

```python
#Check to make sure the data transfer to Employees table was successful

db.execute("SELECT * FROM Employees")
print(db.fetchall())
dbconn.commit()
```
